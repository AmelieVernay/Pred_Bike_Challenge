## Here I try to see if a linear regression to check if weather explains bike traffic intensity would be relevent. (Spoil: no)

I use the dataset `weather` generated by the script in `weather_update.py` and confront it to the dataset `weekatnine` generated by the script `preprocess.py`, containing the number of bike passed each day of the week at 9:00 AM on the totem for the month of March 2021.

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import date
```

```python
weekatnine = pd.read_csv("../Data/marchweekatnine.csv")
weather = pd.read_csv("weather.csv")
```

```python
weather['Date'] = pd.to_datetime(weather['Date'], format='%Y-%m-%d')
weather.set_index('Date', inplace=True)
weather = weather.loc[weather.index.dayofweek <= 4]
weather.reset_index(inplace=True)
```

```python
del weather['Unnamed: 0']
del weather['Date']
del weekatnine['Date']
```

```python
together = pd.concat([weekatnine, weather.reindex(weekatnine.index)], axis=1)
```

```python
together
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=together["wind"], y=together["Today's total"],
                    mode='markers',
                    name='wind (km/h)',
                    marker_size=7,
                    marker_color='rgba(144, 27, 12, .9)'))
fig.add_trace(go.Scatter(x=together["cloud"], y=together["Today's total"],
                    mode='markers',
                    name='cloud (% sky cov.)',
                    marker_size=7,
                    marker_color='rgba(215, 125, 25, .9)'))

fig.update_yaxes(title_text='number of bike around 9:00 AM')
fig.update_layout(title_text='Bike traffic intensity depending on the weather',
                  title_font_size=21,
                  title_x=0.85,
                  title_y=0.88,
                  legend_y=0.93,
                  legend_x=0.83,
                  margin_l=90,
                  legend=dict(bordercolor="Black",
                              borderwidth=1))

fig.layout.template = "plotly_white"
fig.show()
```

The above graph clearly shows that a linear regression wouldn't be justified. No linear trend appears, neither for the wind nor for nebulosity, just a widespread set of points.
